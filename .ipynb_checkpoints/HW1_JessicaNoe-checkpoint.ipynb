## Add column for total East plus West count, check that they were added correctly

```python
df['Total E+W Bike Count'] = df['Fremont Bridge East Sidewalk']+df['Fremont Bridge West Sidewalk']
```

## Convert 'Date' column to a datetime64 type - this will allow me to use built in functions to access the hour and year data in the 'Date' string

```python
time = pd.to_datetime(df['Date'])
```

```python
time.head()
```

```python
type(time) # confirm type
```

```python
type(time[0]) # confirm the type of a single entry
```

```python
time[1670]  # view a single entry
```

```python
time[1670].hour  # check that the function call .hour works on a time entry
```

## Add hour and year columns to the dataframe, check

```python
df['Hour of Day'] = time.dt.hour  # function returns the hour portion of each timestamp element in the 'time' series
```

```python
df.head()  # check that column was added
```

```python
df['Year'] = time.dt.year
```

```python
df.head() # check that column was added
```

## Select just the subset that occurred in 2016

```python
criteria = df['Year'] == 2016
```

```python
#criteria #just checking that it looks correct
```

```python
df_2016 = df[criteria]
```

```python
df_2016  #just checking that it looks correct
```

## Calculate the aggregate trips for 2016 based on hour of the day

```python
df_2016_TotalByHour = df_2016.groupby('Hour of Day').sum()
```

```python
df_2016_TotalByHour
```

## Plot the hourly trips in a bar chart 

```python
df_2016_TotalByHour.plot(kind='bar', y='Total E+W Bike Count')
```

## Busiest hour of the day is 17:00 - 18:00 (5 pm - 6 pm). 
## Average daily number of trips during this hour:

```python
round(max(df_2016_TotalByHour['Total E+W Bike Count'])/366,0)
```
